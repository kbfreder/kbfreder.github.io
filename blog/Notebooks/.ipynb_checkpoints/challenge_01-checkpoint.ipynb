```python
import pandas as pd
import random
import itertools
import datetime as dt
import matplotlib.pyplot as plt
import calendar
import numpy as np
%matplotlib inline
```

## Challenge 01-mta: Benson challenges

#### Challenge 1   
  
- Open up a new IPython notebook
- Download a few MTA turnstile data files
- Open up a file, use csv reader to read it and ensure there is a column for each feature (C/A, UNIT, SCP, STATION). These are the first four columns.

```python
df1 = pd.read_csv('challenges_data/turnstile_180526.txt')
df2 = pd.read_csv('challenges_data/turnstile_180602.txt')
df3 = pd.read_csv('challenges_data/turnstile_180609.txt')
df4 = pd.read_csv('challenges_data/turnstile_180616.txt')
df5 = pd.read_csv('challenges_data/turnstile_180623.txt')
df6 = pd.read_csv('challenges_data/turnstile_180630.txt')
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6])
df.reset_index(inplace=True, drop=True)
df.columns = df.columns.str.strip()
```

### Challenge 2
Let's turn this into a time series.
- Create a new column that specifies the date and time of each entry.

```python
df['DATETIME'] = pd.to_datetime(df.DATE + ' ' + df.TIME, format='%m/%d/%Y %H:%M:%S')
```

### Challenge 3
These counts are for every n hours. (What is n?) 

```python
# n = 4 hours
```

We want total daily entries.
- Calculate to total daily entries for each turnstile.

Your result should be a new data frame that has the same keys, but now we have a single value for a single day, which is the total number of passengers that entered through this turnstile on this day.

*Create key for each station, since SCP #'s are re-used between stations*

```python
df['STATION_KEY'] = df['C/A'] + ' ' + df['UNIT'] + ' ' + df['STATION']
```

*Data in df are counter values. Need to calculate difference to get actual # of entries & exits.*

```python
df.sort_values(['STATION_KEY', 'SCP', 'DATETIME'], inplace=True)
df['ENTRY_DIFFS'] = df.groupby(['STATION_KEY','SCP'])['ENTRIES'].diff(periods=-1)*-1
```

*The majority of 'ENTRY_DIFF' values are reasonable, but there are some large negative and positive values. (There shouldn't, in fact, be any negative values.)*

```python
df['ENTRY_DIFFS'].describe()
```

*If we investigate, these appear to be due to a resetting of the turnstile counters.*

```python
indx = df.loc[(df['ENTRY_DIFFS'] > 2E5)][:1].index[0]
df.loc[indx-2:indx+2]
```

*Filter out these outliers*

```python
# 200,000 entries per 4-hr time period is a reasonable cut-off
df = df[(df['ENTRY_DIFFS'] < 2E5) 
              & (df['ENTRY_DIFFS'] > 0)]
```

*Now, group by Turnstile (SCP) and sum over entries*

```python
df2 = df.groupby(['STATION_KEY','SCP','DATE'])['ENTRY_DIFFS'].sum().reset_index()
df2.rename(columns={'ENTRY_DIFFS':'TOTAL_ENTRIES'}, inplace=True)
```

### Challenge 4
Now plot the daily time series for a turnstile.

```python
df2['DATE'] = pd.to_datetime(df2['DATE'], format="%m/%d/%Y")
```

```python
rand_idx = df2.sample().index.values[0]
rand_stkey = df2.loc[rand_idx, 'STATION_KEY']
rand_tnst = df2.loc[rand_idx, 'SCP']
rand_df = df2[(df2['STATION_KEY'] == rand_stkey) & (df2['SCP'] == rand_tnst)]
```

```python
plt.plot(rand_df['DATE'], rand_df['TOTAL_ENTRIES'])
plt.ylabel('Total Daily Entries')
plt.title(rand_stkey)
plt.xticks(rotation=90);
```

### Challenge 5
So far we've been operating on a single turnstile level, let's combine turnstiles in the same ControlArea/Unit/Station combo. There are some ControlArea/Unit/Station groups that have a single turnstile, but most have multiple turnstilea-- same value for the C/A, UNIT and STATION columns, different values for the SCP column.

We want to combine the numbers together -- for each ControlArea/UNIT/STATION combo, for each day, add the counts from each turnstile belonging to that combo.

```python
df3 = df2.groupby(['STATION_KEY','DATE'])['TOTAL_ENTRIES'].sum().reset_index()
```

```python
df3.head()
```

### Challenge 6

Similarly, combine everything in each station, and come up with a time series of [(date1, count1),(date2,count2),...] type of time series for each STATION, by adding up all the turnstiles in a station.

```python
df4 = df.groupby(['STATION','DATE'])['ENTRY_DIFFS'].sum().reset_index()
df4.rename(columns={'ENTRY_DIFFS':'TOTAL_ENTRIES'}, inplace=True)
```

### Challenge 7

Plot the time series (either daily or your preferred level of granularity) for a station.

```python
rand_idx = df4.sample().index.values[0]
rand_st = df4.loc[rand_idx, 'STATION']
rand_df = df4[(df4['STATION'] == rand_st)]
```

```python
plt.plot(rand_df['DATE'], rand_df['TOTAL_ENTRIES'])
plt.ylabel('Total Daily Entries')
plt.title(rand_st)
xlabels = rand_df[::2]['DATE']
xlocs = np.arange(0, len(rand_df), 2)
plt.xticks(xlocs, xlabels, rotation=90);
```

### Challenge 8

Select a station and find the total daily counts for this station. Then plot those daily counts for each week separately.

To clarify: if I have 10 weeks of data on the 28th st 6 station, I will add 10 lines to the same figure (e.g. running plt.plot(week_count_list) once for each week). Each plot will have 7 points of data.

```python
df4['DATE'] = pd.to_datetime(df4['DATE'])
df4['DAY_OF_WEEK'] = df4['DATE'].dt.weekday
df4['WEEK_NUM'] = df4['DATE'].dt.week
```

```python
rand_idx = df4.sample().index.values[0]
rand_st = df4.loc[rand_idx, 'STATION']
rand_df = df4[(df4['STATION'] == rand_st)]
rand_df.sort_values(['WEEK_NUM', 'DAY_OF_WEEK'], inplace=True)
rand_pt = rand_df.pivot('WEEK_NUM', 'DAY_OF_WEEK','TOTAL_ENTRIES')
```

```python
xlabels = [x[:3] for x in calendar.day_name]

for wknum in rand_pt.index:
    label = dt.datetime.strptime("2018-" + str(wknum-1) + '-0', "%Y-%W-%w").strftime('%m-%d')
    plt.plot(rand_pt.loc[wknum].dropna(), label=label)
plt.title(rand_st)
plt.ylabel('Daily entries')
plt.xticks(np.arange(7), xlabels)
plt.legend(title='Week of');
```

### Challenge 9

Over multiple weeks, sum total ridership for each station and sort them, so you can find out the stations with the highest traffic during the time you investigate

```python
df5 = df4.groupby('STATION')['TOTAL_ENTRIES'].sum().sort_values(ascending=False).reset_index()
```

```python
df5.head()
```

### Challenge 10

Make a single list of these total ridership values and plot it with

    plt.hist(total_ridership_counts)

to get an idea about the distribution of total ridership among different stations.

This should show you that most stations have a small traffic, and the histogram bins for large traffic volumes have small bars.

```python
st_tots = df5['TOTAL_ENTRIES']
```

```python
dmin = df['DATETIME'].dt.date.min().strftime(format='%m-%d-%y')
dmax = df['DATETIME'].dt.date.max().strftime(format='%m-%d-%y')
```

```python
plt.hist(st_tots)
plt.ylabel('# Stations')
plt.xlabel('Total Entries')
plt.title('Total Entries per station from ' + dmin + ' to ' + dmax);
```

If you want to see which stations take the meat of the traffic, you can sort the total ridership counts and make a plt.bar graph.

```python
indices = range(len(st_tots))
plt.barh(indices, st_tots[::-1])
plt.tick_params(left=False, labelleft=False)
plt.xlabel('Total Entries')
plt.title('Total Entries per station from ' + dmin + ' to ' + dmax);
```
